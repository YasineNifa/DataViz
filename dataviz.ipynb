# Data Visualization 

# Voir les differentes parametres 

- Data visualization est la representation graphique des données.
- Il s'agit de produire des images qui communiquent les relations entre les données représentées à l'utilisateur
- La visualisation des données est une partie essentielles de l'analyse des données et de l'apprentissage automatique. Mais choisir le bon type de visualisation est souvent difficile
- Parmis les types de graphes les plus utilisés :
    - Line graph
    - Scatter plot
    - Histogram and Frequency Distribution
    - Heatmap
    - Contour Plot
    - Box Plot
    - Bar Chart
- Libraries utilisées sont :
    - matplotlib : la bibliothéque la plus ancienne de Plotting et visualisation 
    - seaborn : Une bibliothèque de visualisation facile à utiliser qui s'appuie sur Matplotlib et vous permet de créer de superbes graphiques avec seulement quelques lignes de code.

```python
# Import libraries
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

### Line Chart
- affiche les informations sous la forme d'une serie de points de données ou de marqueurs, reliés par des lignes droites.
- On peut personnaliser la forme, la taille, la couleur et d'autres éléments esthétiques des marqueurs et des lignes pour une meilleure clarté visuelle

#### Exemple
- We'll create a line chart to compare the yields of apples and oranges over 12 years in the imaginary region of Hoenn.

```python
# Sample data
years = np.arange(2000,2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]

# We can add new params to the plot functions
    # linewidth,markersize,markeredgewidth, markeredgecolor
    # linewidth = 4,markeredgewidth = 4,
plt.plot(years,apples,'bx-',label = "Apples", markersize = 12,markeredgecolor = 'black')
plt.plot(years,oranges,'ro--',label = "Oranges")
plt.title("Crop Yields in Hoenn Region")
plt.xlabel("Year")
plt.ylabel("Yield(tons)")
plt.legend()
```

### Scatter Plot
- (nuage de points) : les valeurs de 2 variables sont representées sous forme de points sur une grille à 2 dimensions
- On peut utiliser une troisième variable qui permet de determiner la taille ou la couleur des points

#### Exemple

```python
# The iris dataset is included with the seaborn library
data = sns.load_dataset('iris')
data.sample(5)
```

```python
plt.scatter(data.sepal_length,data.sepal_width)
plt.title("Flowers")
```

```python
# create a scatter chart using seaborn
sns.scatterplot(data.sepal_length,data.sepal_width,hue=data.species,s=100)
plt.title("Flowers")
```

### Heatmap
- Un heatamp est utilisé pour visualiser des données bidimensionnelles comme une matrice ou un tableau en utilisant des couleurs.

#### Exemple :
- We'll use another sample dataset from Seaborn, called "flights", to visualize monthly passenger footfall at an airport over 12 years.

```python
flights = sns.load_dataset('flights').pivot("month", "year", "passengers")
flights.sample(5)
```

```python
flights = sns.load_dataset("flights").pivot("month", "year", "passengers")

# Chart Title


# Draw a heatmap with the numeric values in each cell
#linewidths=.5, 
g = sns.heatmap(flights, 
            fmt="d", 
            linewidths=1.,
            annot=True, 
            cmap='Blues',
            annot_kws={"fontsize":13});
plt.xticks(rotation=90)
#plt.title("No. of Passengers (1000s)")
# The botton and the top of the plot are cut off
# Problem related to matplotlib version you have to upgrade. But we can do this
# fix for mpl bug that cuts off top/bottom of seaborn viz

b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show() # ta-da!
```

### Contour Plot
- nous aide à visualiser 3 variables numériques en deux dimensions. Une variable est représentée sur l'axe horizontal et une seconde variable est représentée sur l'axe vertical. La troisième variable est représentée par un dégradé de couleur et des isolignes (lignes de valeur constante), il represente la densité des valeurs dans une region de graphe

#### Exemple :
- We can visulize the values of sepal width & sepal length from the flowers dataset using a contour plot. The shade of blue represent the density of values in a region of the graph.

```python
sns.kdeplot(data.sepal_length,data.sepal_width,shade = True,shade_lowest=False)
plt.title("Flowers")
```

### Box Plot
- il montre la distribution de données en utilisant des box "boites" et des "whiskers". l'extrémité inférieure de la boite représente le 1er quartile(25% des valeurs sont en dessous et l'extremité supérieur représente le 3em quartile (c'est à dire que 25% des valeurs sont au dessus. La valeur mediane est representée par une ligneà l'intérieur de la box. whiskers representent représentent les valeurs minimales et maximales (à l'exclusion parfois des valeurs aberrantes, qui sont représentées sous forme de points).

- medianne : est une valeur x qui permet de couper l'ensemble des valeurs en deux parties égales : mettant d'un côté une moitié des valeurs, qui sont toutes inférieures ou égales à x et de l'autre côté l'autre moitié des valeurs, qui sont toutes supérieures ou égales à x 

#### Exemple :
- We'll use another sample dataset included with Seaborn, called "tips". The dataset contains information about the sex, time of day, total bill and tip amount for customers visiting a restraurant over a week.

```python
tips = sns.load_dataset('tips')
tips.sample(5)
```

- We can use a box plot to visualize the distribution of total bill for each day of the week, segmented by whether the customer was a smoker.

```python
sns.boxplot(tips.day,tips.total_bill,hue = tips.smoker)
```

##  Visualizing Statistical relationships
- L'analyse statistique est un processus visant à comprendre comment les variables d'un ensemble de données sont liées les unes aux autres et comment ces relations dépendent d'autres variables
- Permet de visualiser les tendances et la relation entre les variables
- relplot(), FacetGrid, scatterplot,lineplot

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
#sns.set_theme(style="darkgrid")
```

### Emphasizing continuity with scatter plots

#### scatterplot
- Il est un pilier de la visualisation statique, il décrit la distribution conjointe entre deux variables à l'aide d'un ensemble de points ou chaque point represente une observation dans l'ensemble de données. Cette representation permet à l'oeil de déduire l'existance ou non d'une relation significative entre eux

- Alors que les points sont tracés en deux dimensions, une autre dimension peut être ajoutée au tracé en colorant les points selon une troisième variable. Dans seaborn, cela s'appelle l'utilisation d'une «sémantique de teinte» `hue` semantic", car la couleur du point prend du sens 

- Pour souligner la différence entre les classes et améliorer l'accessibilité, vous pouvez utiliser un style de marqueur différent pour chaque classe: `style`

- `hue` avec une variable catogirielle on applique ce qu'on appelle `qualitative palette`

- `hue` avec une variable numeric on applique ce qu'on appelle `sequential palette`

- `size` changer la taille de chaque point

- `sizes`: normaliser les area of the point into a specific range

```python
tips = sns.load_dataset("tips")
tips.sample(5)
```

```python
sns.relplot(tips.total_bill,tips.tip,kind='scatter')
```

```python
sns.scatterplot(tips.total_bill,tips.tip)
```

```python
sns.scatterplot('total_bill','tip',data=tips)
```

```python
sns.scatterplot('total_bill','tip',hue='smoker',style='smoker',data=tips)
```

```python
sns.scatterplot('total_bill','tip',hue='smoker',style='time',data=tips)
```

```python
sns.scatterplot('total_bill','tip',hue='size',size = 'size', sizes=(15, 200),data=tips)
```

### Aggregation and representing uncertainty
- Les ensembles de données plus complexes auront plusieurs mesures pour la même valeur de la variable x. Le comportement par défaut dans seaborn est d'agréger les multiples mesures à chaque valeur x en traçant la moyenne et l'intervalle de confiance à 95% autour de la moyenne

- Les intervalles de confiance sont calculés à l'aide du bootstrap, qui peut prendre beaucoup de temps pour des ensembles de données plus volumineux. Il est donc possible de les désactiver `ci = None`

- Une autre bonne option, en particulier avec des données plus volumineuses, consiste à représenter l'étalement de la distribution à chaque point temporel en traçant l'écart type au lieu d'un intervalle de confiance `ci = 'sd'`

- Pour désactiver complètement l'agrégation, `estimator = None` définissez le paramètre de l'estimateur sur Aucun. Cela peut produire un effet étrange lorsque les données comportent plusieurs observations à chaque point.

```python
fmri = sns.load_dataset('fmri')
fmri.sample(15)
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line")
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",ci=None)
```

```python
sns.relplot(x="timepoint",y="signal",data=fmri,kind="line",ci='sd')
```

```python
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri)
```

```python
fmri.sample(5)
```

```python
sns.relplot(x="timepoint",y="signal",style="event",hue="region",kind="line",data=fmri)
```

### Plotting with date data
- line plot : visualiser les données associées aux dates et heures réelles. Ces fonctions transmettent les données dans leur format d'origine aux fonctions matplotlib sous-jacentes, et ainsi elles peuvent tirer parti de la capacité de matplotlib à formater les dates dans les ticklabels. Mais tout ce formatage devra avoir lieu au niveau de la couche matplotlib, et vous devriez vous référer à la documentation matplotlib pour voir comment cela fonctionne

- `pd.date_range("2017-01-01",periods=500)`

```python
pd.date_range("2017-01-01",periods=500)
```

```python
df = pd.DataFrame(dict(time = pd.date_range("2017-01-01", periods = 500), value = np.random.randn(500).cumsum()))
g = sns.relplot(x="time",y="value",data=df,kind = "line")
g.fig.autofmt_xdate()
# np.cumsum : Return the cumulative sum of the elements along a given axis.
```

### Showing multiple relationships with facets
- R(v1,EnsVar) and R(v2,EnsVar)

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",col="time" ,data = tips)
```

```python
sns.relplot(x="timepoint", y="signal", hue="subject",
            col="region", row="event", height=3,
            kind="line", estimator=None, data=fmri);
```

- Lorsque vous souhaitez examiner les effets sur de nombreux niveaux d'une variable, il peut être judicieux de mettre en facettes cette variable sur les colonnes, puis d'envelopper les facettes dans les lignes:

```python
#height=3,style="event",col_wrap=5,aspect=.75, linewidth=2.5,
sns.relplot(height=3,aspect=.75,linewidth=2.5,x="timepoint", y="signal", hue="event", col="subject", kind="line",col_wrap=5,data=fmri.query("region == 'frontal'"));
```
